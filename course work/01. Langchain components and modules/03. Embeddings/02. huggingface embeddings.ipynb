```python
# sentence transformer on huggingface embedding technique

from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = "This is a test document"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embeddings.embed_documents([text, "This is another test document"])
doc_result
```
