## Let's use Proprietary LLM from - OpenAI

```python
# !pip install Openai
```

```python
import os
os.environ['OPENAI_API_KEY'] = ''
```

```python
from langchain.llms import OpenAI
```

```python
llm = OpenAI(model_name="gpt-3.5-turbo-instruct")
```

```python
our_query = "What is the currency of India?"
completion = llm(our_query)  # it is a paid model so will throw an error asking us to pay the bill
```

```python
print(completion) # this will print the answer generated by our model
```

## Let's use open-source LLM hosted on Hugging Face

```python
import os
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "" 
# keep in mind to keep the key same as its used by langchain
```

```python
from langchain.llms import HuggingFaceHub
```

```python
llm = HuggingFaceHub(repo_id = "google/flan-t5-large")
```

```python
# the llm takes a prompt as an input and outputs a completion
our_query = "What is the currency of India?"
completion = llm(our_query)
```

```python
print(completion)
```
